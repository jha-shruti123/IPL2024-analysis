```python
player_data = []
```

```python
# prompt: do the above for every single divs and image of the class and append in a list and return

import requests
from bs4 import BeautifulSoup

# URL of the webpage
url_list=["https://www.iplt20.com/teams/chennai-super-kings","https://www.iplt20.com/teams/delhi-capitals","https://www.iplt20.com/teams/gujarat-titans","https://www.iplt20.com/teams/kolkata-knight-riders",
          "https://www.iplt20.com/teams/lucknow-super-giants","https://www.iplt20.com/teams/mumbai-indians","https://www.iplt20.com/teams/punjab-kings","https://www.iplt20.com/teams/rajasthan-royals",
          "https://www.iplt20.com/teams/royal-challengers-bangalore","https://www.iplt20.com/teams/sunrisers-hyderabad"]

for url in url_list:

# Send a GET request to the webpage
 response = requests.get(url)

# Parse the HTML content of the page
 soup = BeautifulSoup(response.content, 'html.parser')
 team_name = soup.find('div', class_='vn-teamTitleODet ps-4 ms-4').find('h2').text.strip()


# Find all divs with class "ih-p-img"
 player_divs = soup.find_all('div', class_='ih-p-img')



# Iterate over the player divs and extract information
 for div in player_divs:
  image_link = div.find('img')['data-src']
  player_name = div.find('h2').text
  role = div.find('span').text.strip()
  player_data.append({"name": player_name, "image": image_link, "team":team_name,"role": role})


print(player_data)
```

```python
# prompt: convert player data into pandas then into a csv file

import pandas as pd

df = pd.DataFrame(player_data)
df.to_csv('player_data.csv', index=False)
df.tail()
```
